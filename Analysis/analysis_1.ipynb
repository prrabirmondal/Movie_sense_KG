```python
import os
import pandas as pd
```

# For KG

```python
root_path = "/mnt/Data/prabirmondal/prabir/python_program/movie_sense/SRI_KG/Movie_sense_KG/Movie_sense_KG/Dataset/Nodes_Edges/Bollywood"

Categories = os.listdir(root_path)
```

### Count duplicate triplets

```python
for category in Categories:
    duplicate_rows = 0
    total_triplet_count = 0
    category_path = os.path.join(root_path, category)
    node_relation_files = os.listdir(category_path)
    # print(len(node_relation_files))
    for node_relation_file in node_relation_files:
        file_pd = pd.read_csv(os.path.join(category_path, node_relation_file))
        total_triplet_count += len(file_pd)
        dups = file_pd.duplicated()
        if dups.sum()>0:
            # print(f'Category = {category}, File_name = {node_relation_file}, Duplicate_rows_count = {dups.sum()}/{len(file_pd)}')
            duplicate_rows += dups.sum()
    duplicate_percentage = (duplicate_rows/total_triplet_count) * 100
    print(f'Duplicate Percentage in Category = {category} = {duplicate_percentage}')
```

```python
aspect_list = ['Accolades',
'Cast',
'Guardians of the Galaxy Vol',
'Music',
'Plot',
'Production',
'Soundtrack',
'Themes',
'summary']


for category in Categories:
    triplet_count = {'Accolades' : 0, 'Cast' : 0, 'Music' : 0, 'Plot' : 0, 'Production' : 0, 'Soundtrack' : 0, 'Themes' : 0, 'summary' : 0}
    total_triplets = 0
    duplicate_triplets = 0
    movie_files = os.listdir(os.path.join(root_path, category))
    for movie_file in movie_files:
        file_path = os.path.join(root_path, category, movie_file)  
        
        try:
            pd_file = pd.read_csv(file_path)
            aspect = movie_file.split('.')[0].split('_')[-1]
            if aspect in triplet_count.keys():                    
                total_triplet = len(pd_file)
                triplet_count[aspect] += total_triplet
                total_triplets += total_triplet
                
                dups = pd_file.duplicated()
                duplicate_triplets += dups.sum()
                # if dups.sum() > 0:
                #     print(f'catetory = {category}, hop = {hop}, question_type = {question_type}, file name = {file_name}')
                #     print(dups)
        except: continue
        
    print(f'Triplet counts for {category}:')
    print("------------------------------------------------------") 
    print(triplet_count)
    print("\n\n")
    
    print(f'Total Triplets in {category} category is : {total_triplets}')
    print(f'Total duplicate questions in {category} category is : {duplicate_triplets}')
    print(f'percentage of duplicate questions = {(duplicate_triplets/total_triplets)*100}')
    print('============================================================\n\n')
        
```

# For Questions

### count questions

```python
root_path = "/mnt/Data/prabirmondal/prabir/python_program/movie_sense/SRI_KG/Movie_sense_KG/Movie_sense_KG/Dataset/Questions_Answers/Bollywood"

Categories = os.listdir(root_path)
```

```python
aspect_list = ['Accolades',
'Cast',
'Guardians of the Galaxy Vol',
'Music',
'Plot',
'Production',
'Soundtrack',
'Themes',
'summary']


for category in Categories:
    total_questions = 0
    duplicate_questions = 0
    hops = os.listdir(os.path.join(root_path, category))
    for hop in hops:
        question_types = os.listdir(os.path.join(root_path, category, hop))
        # print(question_types)
        for question_type in question_types:           
            
            question_count = {'Accolades' : 0, 'Cast' : 0, 'Music' : 0, 'Plot' : 0, 'Production' : 0, 'Soundtrack' : 0, 'Themes' : 0, 'summary' : 0}
            
            file_names = os.listdir(os.path.join(root_path, category, hop, question_type))
            for file_name in file_names:
                file_path = os.path.join(root_path, category, hop, question_type, file_name)
                try:
                    pd_file = pd.read_csv(file_path)
                    aspect = file_name.split('.')[0].split('_')[-1]
                    if aspect in question_count.keys():                    
                        total_question = len(pd_file)
                        question_count[aspect] += total_question
                        total_questions += total_question
                        
                        dups = pd_file.duplicated()
                        duplicate_questions += dups.sum()
                        # if dups.sum() > 0:
                        #     print(f'catetory = {category}, hop = {hop}, question_type = {question_type}, file name = {file_name}')
                        #     print(dups)
                except: continue
        
            print(f'Question count for {category}/{hop}/{question_type}:')
            print("------------------------------------------------------") 
            print(question_count)
            print("\n\n")
        
    print(f'Total question in {category} category is : {total_questions}')
    print(f'Total duplicate questions in {category} category is : {duplicate_questions}')
    print(f'percentage of duplicate questions = {(duplicate_questions/total_questions)*100}')
    print('============================================================\n\n')
```
